```python
df = pd.get_dummies(data, columns=['Geography', 'Gender'], drop_first=True)
```

```python
df.drop(columns=["RowNumber","CustomerId","Surname"],inplace=True)
df.head()
```

```python
# Convert boolean columns to integers
# Assuming 'col1' and 'col2' are the columns with True/False values
df['Geography_Germany'] = df['Geography_Germany'].astype(int)
df['Geography_Spain'] = df['Geography_Spain'].astype(int)
df['Gender_Male'] = df['Gender_Male'].astype(int)
```

```python
df.head()
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited'].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```
