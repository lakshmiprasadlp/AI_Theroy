```python
import boto3

# Replace 'your_access_key', 'your_secret_key', and 'your_region' with your AWS credentials and region

# Replace 'YourTableName' with the desired table name
table_name = 'YourTableName'

# Create a DynamoDB resource
dynamodb = boto3.resource('dynamodb', region_name=aws_region, aws_access_key_id=aws_access_key, aws_secret_access_key=aws_secret_key)

# Define the table schema
table = dynamodb.create_table(
    TableName=table_name,
    KeySchema=[
        {
            'AttributeName': 'PrimaryKey',  # Replace with your primary key attribute name
            'KeyType': 'HASH'  # HASH for hash key, RANGE for range key
        }
        # Add more key schema entries for composite keys if needed
    ],
    AttributeDefinitions=[
        {
            'AttributeName': 'PrimaryKey',  # Replace with your primary key attribute name
            'AttributeType': 'S'  # S for String, N for Number
        }
        # Add more attribute definitions for other attributes
    ],
    ProvisionedThroughput={
        'ReadCapacityUnits': 5,  # Adjust based on your read capacity needs
        'WriteCapacityUnits': 5  # Adjust based on your write capacity needs
    }
)

# Wait for the table to be created before moving forward
table.meta.client.get_waiter('table_exists').wait(TableName=table_name)

print(f"Table '{table_name}' created successfully.")
```
